```python
import pandas as pd
import numpy as np
import requests
```

Lấy Danh sách sản phẩm

```python

# URL API của Unwrangle
url = "https://data.unwrangle.com/api/getter/"

# API key
api_key = "c740d847c201b46a93f7c5135186e0f2c496df88"

# Hàm để lấy dữ liệu từ API cho từng trang
def get_data_from_api(page):
    params = {
        "platform": "amazon_search",  # Nền tảng Amazon với kiểu dữ liệu tìm kiếm
        "search": "robot+vacuum+cleaner",  # Từ khóa tìm kiếm
        "country_code": "us",  # Quốc gia USA
        "page": page,  # Số trang
        "api_key": api_key  # API Key của bạn
    }
    response = requests.get(url, params=params)
    
    # Kiểm tra trạng thái yêu cầu
    if response.status_code == 200:
        return response.json()  # Trả về dữ liệu JSON
    else:
        print(f"Yêu cầu không thành công, mã lỗi: {response.status_code}")
        return None

# Thu thập dữ liệu từ 5 trang
all_data = []
for page in range(1, 50):
    data = get_data_from_api(page)
    if data:
        all_data.extend(data['results'])  # Thêm dữ liệu từ mỗi trang vào danh sách

# Chuyển dữ liệu thành DataFrame
df = pd.DataFrame(all_data)

# Hiển thị DataFrame trong Streamlit

# Lưu DataFrame vào file CSV
csv_file_path = 'amazon_products.csv'  # Đường dẫn file CSV, bạn có thể thay đổi nếu cần
df.to_csv(csv_file_path, index=False)
```

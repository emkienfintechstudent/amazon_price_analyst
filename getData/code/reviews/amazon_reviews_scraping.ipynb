```python
# T√™n file CSV
output_file = "amazon_reviews_top_products.csv"

# M·ªü file CSV ·ªü ch·∫ø ƒë·ªô ghi ti·∫øp (append)
with open(output_file, mode='a', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)

    # N·∫øu file r·ªóng, th√™m ti√™u ƒë·ªÅ c·ªôt
    if file.tell() == 0:  # Ki·ªÉm tra n·∫øu file r·ªóng
        writer.writerow(["ASIN","Rating", "Title", "Body", "Date"])

    # V√≤ng l·∫∑p ƒë·ªÉ duy·ªát qua c√°c trang v√† thu th·∫≠p d·ªØ li·ªáu
    while True:
        try:
            # V√≤ng l·∫∑p qua c√°c ƒë√°nh gi√° tr√™n trang
            for i in range(1, 11):  # ƒêi·ªÅu ch·ªânh ph·∫°m vi n·∫øu c√≥ nhi·ªÅu ho·∫∑c √≠t h∆°n 10 ƒë√°nh gi√°
                try:
                    # L·∫•y ti√™u ƒë·ªÅ
                    title = driver.find_element(By.XPATH, f'/html/body/div[1]/div[3]/div/div[1]/div/div[1]/div[5]/div[3]/div/div[{i+1}]/div/div/div[2]/a/span[2]')
                    
                    # L·∫•y n·ªôi dung v√† lo·∫°i b·ªè d·∫•u xu·ªëng d√≤ng
                    body = driver.find_element(By.XPATH, f'/html/body/div[1]/div[3]/div/div[1]/div/div[1]/div[5]/div[3]/div/div[{i+1}]/div/div/div[4]').text
                    body = body.replace("\n", " ").replace("\r", " ")  # Lo·∫°i b·ªè d·∫•u xu·ªëng d√≤ng v√† thay b·∫±ng kho·∫£ng tr·∫Øng
                    body = " ".join(body.split())  # Lo·∫°i b·ªè kho·∫£ng tr·∫Øng th·ª´a

                    # L·∫•y ng√†y
                    date = driver.find_element(By.XPATH, f'/html/body/div[1]/div[3]/div/div[1]/div/div[1]/div[5]/div[3]/div/div[{i+1}]/div/div/span')

                    # Ghi v√†o file CSV, ƒë√°nh gi√° m·∫∑c ƒë·ªãnh l√† 1 (b·∫°n c√≥ th·ªÉ thay ƒë·ªïi n·∫øu c·∫ßn)
                    writer.writerow(['B0D9W3B622',1, title.text, body, date.text])
                    print(f"Title: {title.text}")
                    print(f"Body: {body}")
                except NoSuchElementException:
                    print(f"Kh√¥ng t√¨m th·∫•y ƒë√°nh gi√° t·∫°i v·ªã tr√≠ {i} tr√™n trang.")
            
            # Cu·ªôn xu·ªëng cu·ªëi trang ƒë·ªÉ t·∫£i th√™m ƒë√°nh gi√°
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)  # Ch·ªù t·∫£i n·ªôi dung
            
            # Nh·∫•p v√†o n√∫t "Next" n·∫øu c√≥
            try:
                next_page = driver.find_element(By.XPATH, '//*[@id="cm_cr-pagination_bar"]/ul/li[2]/a')
                next_page.click()
                time.sleep(3)  # Ch·ªù trang ti·∫øp theo t·∫£i
            except NoSuchElementException:
                print("Kh√¥ng t√¨m th·∫•y n√∫t Next ho·∫∑c kh√¥ng c√≤n trang n√†o.")
                break

        except Exception as e:
            print(f"L·ªói x·∫£y ra: {e}")
            break
```

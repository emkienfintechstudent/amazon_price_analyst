```python
# Tên file CSV
output_file = "amazon_reviews_top_products.csv"

# Mở file CSV ở chế độ ghi tiếp (append)
with open(output_file, mode='a', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)

    # Nếu file rỗng, thêm tiêu đề cột
    if file.tell() == 0:  # Kiểm tra nếu file rỗng
        writer.writerow(["ASIN","Rating", "Title", "Body", "Date"])

    # Vòng lặp để duyệt qua các trang và thu thập dữ liệu
    while True:
        try:
            # Vòng lặp qua các đánh giá trên trang
            for i in range(1, 11):  # Điều chỉnh phạm vi nếu có nhiều hoặc ít hơn 10 đánh giá
                try:
                    # Lấy tiêu đề
                    title = driver.find_element(By.XPATH, f'/html/body/div[1]/div[3]/div/div[1]/div/div[1]/div[5]/div[3]/div/div[{i+1}]/div/div/div[2]/a/span[2]')
                    
                    # Lấy nội dung và loại bỏ dấu xuống dòng
                    body = driver.find_element(By.XPATH, f'/html/body/div[1]/div[3]/div/div[1]/div/div[1]/div[5]/div[3]/div/div[{i+1}]/div/div/div[4]').text
                    body = body.replace("\n", " ").replace("\r", " ")  # Loại bỏ dấu xuống dòng và thay bằng khoảng trắng
                    body = " ".join(body.split())  # Loại bỏ khoảng trắng thừa

                    # Lấy ngày
                    date = driver.find_element(By.XPATH, f'/html/body/div[1]/div[3]/div/div[1]/div/div[1]/div[5]/div[3]/div/div[{i+1}]/div/div/span')

                    # Ghi vào file CSV, đánh giá mặc định là 1 (bạn có thể thay đổi nếu cần)
                    writer.writerow(['B0D2XXSPCR',5, title.text, body, date.text])
                    print(f"Title: {title.text}")
        
                    print(f"Body: {body}")
                except NoSuchElementException:
                    print(f"Không tìm thấy đánh giá tại vị trí {i} trên trang.")
            
            # Cuộn xuống cuối trang để tải thêm đánh giá
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)  # Chờ tải nội dung
            
            # Nhấp vào nút "Next" nếu có
            try:
                next_page = driver.find_element(By.XPATH, '//*[@id="cm_cr-pagination_bar"]/ul/li[2]/a')
                next_page.click()
                time.sleep(3)  # Chờ trang tiếp theo tải
            except NoSuchElementException:
                print("Không tìm thấy nút Next hoặc không còn trang nào.")
                break

        except Exception as e:
            print(f"Lỗi xảy ra: {e}")
            break
```

## extract date

```python
df = pd.read_csv('../../data/amazon_reviews_remove_space_remove_duplicate.csv')
df
```

```python
# Hàm để trích xuất ngày tháng năm từ chuỗi
def extract_date(date_str):
    # Tìm phần "on" trong chuỗi và cắt phần sau từ đó
    date_part = date_str.split("on")[-1].strip()
    # Chuyển đổi chuỗi ngày tháng thành kiểu datetime
    return pd.to_datetime(date_part, errors='coerce')

# Áp dụng hàm cho cột "Date"
df['Extracted_Date'] = df['Date'].apply(extract_date)
```

```python
# Trích xuất ngày, tháng, năm từ cột đã chuyển đổi
df['Year'] = df['Extracted_Date'].dt.year
df['Month'] = df['Extracted_Date'].dt.month
df['Day'] = df['Extracted_Date'].dt.day
```

```python
df.to_csv('../../data/amazon_reviews_remove_space_remove_duplicate.csv', index=False, encoding='utf-8')
```
